```python
#FUNCTIONS

def count_plot(x,hue,title):
    plt.figure(figsize=(18,4))
    sns.countplot(x=df[x], hue=df[hue])
    plt.title(f'{title} {x} by {hue}', fontsize=16, fontweight='bold')
    plt.tick_params(axis='x', labelrotation=90, labelsize=12)
    plt.xlabel(x, fontsize=14)
    return plt.show()



def bar_plot(x,y,title):
    plt.figure(figsize=(18,4))
    sns.barplot(x=df[x], y=df[y])
    plt.title(title, fontsize=16, fontweight='bold')
    plt.tick_params(axis='x',rotation=90,labelsize=12)
    plt.tick_params(axis='y',labelsize=12)
    plt.xlabel(x, fontsize=14)
    plt.ylabel(y, fontsize=14)
    return plt.show()


def scatter_plot(x,y,title):
    plt.figure(figsize=(18,4))
    sns.scatterplot(x=df[x], y=df[y])
    plt.title(title, fontsize=16, fontweight='bold')
    plt.tick_params(axis='x',labelsize=12)
    plt.tick_params(axis='y',labelsize=12)
    plt.xlabel(x, fontsize=14)
    plt.ylabel(y, fontsize=14)
    return plt.show()
```

# Data Spliting

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn import metrics
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
#Encoding a DataFrame typically involves converting categorical variables into numerical representations
le = LabelEncoder()

cat_columns = df.select_dtypes(include='O').columns

df[cat_columns] = df[cat_columns].apply(lambda col: le.fit_transform(col))
```

```python
df
```

```python
X= df.drop('Owner',axis=1)
Y = df['Owner']

X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=36)
```

#Model Buliding

```python
model1 = LogisticRegression()
model1.fit(X_train, Y_train)
y_pred1 = model1.predict(X_test)

accuracy_reg = accuracy_score(Y_test, y_pred1)
conf_matrix_reg = confusion_matrix(Y_test, y_pred1)
classification_rep_reg = classification_report(Y_test, y_pred1)

print(f"Accuracy: {accuracy_reg:.2f}")
print()
print("Confusion Matrix:")
print(conf_matrix_reg)
print()
print("Classification Report:")
print(classification_rep_reg)
```

```python
model2= RandomForestClassifier()
model2.fit(X_train, Y_train)
y_pred2 = model2.predict(X_test)

accuracy_rf = accuracy_score(Y_test, y_pred2)
conf_matrix_rf = confusion_matrix(Y_test, y_pred2)
classification_rep_rf = classification_report(Y_test, y_pred2)

# Print the results
print(f"Accuracy: {accuracy_rf:.2f}")
print()
print("Confusion Matrix:")
print(conf_matrix_rf)
print()
print("Classification Report:")
print(classification_rep_rf)
```

```python
model3= GradientBoostingClassifier()
model3.fit(X_train, Y_train)
y_pred3 = model3.predict(X_test)

accuracy_gb = accuracy_score(Y_test, y_pred3)
conf_matrix_gb= confusion_matrix(Y_test, y_pred3)
classification_rep_gb = classification_report(Y_test, y_pred3)

# Print the results
print(f"Accuracy: {accuracy_gb:.2f}")
print()
print("Confusion Matrix:")
print(conf_matrix_gb)
print()
print("Classification Report:")
print(classification_rep_gb)
```

#Results

```python
print(f"Accuracy score (Logistic Regression): {accuracy_reg:.2f}")
print(f"Accuracy score (Random Forest): {accuracy_rf:.2f}")
print(f"Accuracy score (Gradient Boosting Classifier): {accuracy_gb:.2f}")
```

```python
print("Confusion Matrix for Logistic_Regression")
print(conf_matrix_reg)
print()
print("Confusion Matrix for Random Forest")
print(conf_matrix_rf)
print()
print("Confusion Matrix for Gradient Boosting Classifier")
print(conf_matrix_gb)
print()
```

```python
models = ['Logistic Regression', 'Random Forest','Gradient Boosting Classifier', ]
accuracies = [accuracy_reg*100, accuracy_rf*100, accuracy_gb*100]

plt.bar(models, accuracies, color=['green','orange','blue'])
plt.xlabel("Models")
plt.ylabel("Accuracy")
plt.title("Model Comparison - Accuracy")
plt.ylim([0, 100])
plt.show()
```
